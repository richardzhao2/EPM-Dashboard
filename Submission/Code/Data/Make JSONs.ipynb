```python
import pandas as pd
import json
```

```python
players = pd.read_csv('Files/Player_Map.csv')
```

```python
players.head()
```

```python
players['Full Name'] = players['First_Name'].map(str) + players['Last_Name']
```

```python
players.head()
```

```python
players['SV_Global_Player_ID'] = players['SV_Global_Player_ID'].astype('int').astype('str')
players['Person_ID'] = players['Person_ID'].astype('int').astype('str')
```

```python
players['Merged'] = [[val1, val2] for val1, key, val2 in zip(players['Person_ID'], players['SV_Global_Player_ID'], players['Full Name'])]
```

```python
playersDict = dict(zip(players['SV_Global_Player_ID'], players['Merged']))
print(playersDict)
```

```python
players['Merged']
```

```python

tempFile = open('SPORTVU UTIL/SVNames.json', 'w')
tempFile.write(json.dumps(playersDict))
tempFile.close()
```
